```python
""""""""""""""""""
""" IMPORTS """
""""""""""""""""""

from drone_home_model import run_sim, DroneHomeModel
import agentpy as ap
import matplotlib.pyplot as plt
import numpy as np

def run_exp(model, parameters, runs, random=True):
    """
    Run experiment for a set of parameters
    """
    # Get 10 different random seeds 
    sample = ap.Sample(parameters, runs, randomize=random)
    # Run experiment and gather results
    exp = ap.Experiment(model, sample, iterations = 50, record=True)
    results = exp.run(-1, display=True)
    return results


""""""""""""""""""
""" RUN SIM """
""""""""""""""""""

sim_parameters = {
    'size': 300,
    'steps': 10000,
    'population' : ap.IntRange(1,10),
    'dispersion_radius' : 25,
    'camera_radius' : 5,
    'random_walk_strength' : 1,
    'random_walk_delay' : 100,
    'border_strength' : 0.5,
    'sborder_strength' : 0.5,
    'dispersion_strength' : 5,
    'drone_speed' : 0.3
}

# Run experiment and collect data for analysis
results = run_exp(DroneHomeModel, sim_parameters, 10)
```

```python
import numpy as np

print(np.split(np.array(results.reporters["time_to_finding"]), 50))
```

```python
fig, ax = plt.subplots(figsize=(8,6), dpi=1200)

fig.patch.set_facecolor('#004AAD')
ax.set_facecolor('#004AAD')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')
ax.spines['left'].set_color('white')

c='white'

times = np.split(np.array(results.reporters["time_to_finding"]), 10)
ax.boxplot(times, showfliers=False, patch_artist=True,
           boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color='orange', linewidth='2'),)

ax.set(xlabel="Number of drones in fleet", ylabel="Time to locate person")

plt.show()
```
